## CSV (of all USA Zip Codes with their Latitudes and Longitudes
A csv file AWA comma-separated values

```python
import pandas as pd

zip_df = pd.read_csv("data/free-zipcode-database-Primary.csv",index_col=0)
zip_df
```

We can use the pandas `.loc` attribute to look up the zip code of the school and the corresponding latitude and longitude.

```python
zip_df.loc[90034, ['Lat', 'Long']]
```

```python
zip_count = {}

for zip_code in unique_zips:
    lat = zip_df.loc[zip_code]['Lat']
    long = zip_df.loc[zip_code]['Long']
    zip_count[str(zip_code)] = [zip_codes.count(zip_code), lat, long]
```

```python
zip_count
```

```python
gala_map = folium.Map(location=[gala_lat - .05, gala_long],
                        zoom_start=11, max_zoom=15, min_zoom=8,
                        tiles="CartoDB positron")

for zip_code in zip_count:
    folium.CircleMarker(location=[zip_count[zip_code][1], zip_count[zip_code][2]], 
                        fill=True, popup= str(zip_code) + '<br>' + str(zip_count[zip_code][0]) + '\n Students', 
                        color='#65318f', radius = zip_count[zip_code][0], 
                        fill_opacity = 1 / zip_count[zip_code][0]).add_to(gala_map)
    
folium.Marker(location=[gala_lat, gala_long],
             icon=folium.Icon(color='black',icon='graduation-cap', prefix='fa'), 
              popup = GALA_Address).add_to(gala_map)

#folium.CircleMarker(location=[gala_lat, gala_long], radius = 10, color = '#65318f').add_to(gala_map)

gala_map
```
