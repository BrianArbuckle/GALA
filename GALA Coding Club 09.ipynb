## Geographic Coordinate System

A geographic coordinate system is a coordinate system used in geography that enables every location on Earth to be specified by a set of numbers, letters or symbols. The coordinates are often chosen such that one of the numbers represents a vertical position, and two or three of the numbers represent a horizontal position. A common choice of coordinates is **latitude, longitude** and **elevation**.

For a 2d map, all we need is latitude, longitude.

## Latitude

Is the angular distance of the places in North and South direction of Earth equator.

All parallel circles from the equator up to the poles are called parallels of latitudes. Latitudes are measured in degrees.  Equator is the 0° of latitude.  Everything north to the north pole is a positive number up to 90, while everything south is a negative number.

## Longitude

Is the angular distance of the places in East and West direction of the Prime Meridian, which is in Greenwich London, England.  Prime Meridian, the 0° of longitude.

## converting zip codes to latitude and longitude using geopy

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="gala")

from geopy.extra.rate_limiter import RateLimiter
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
import time

GALA_Address = '1067 West Blvd, Los Angeles, CA 90019'

location = geolocator.geocode(GALA_Address)
print(location.address)
print(location.latitude, location.longitude)

gala_lat = location.latitude
gala_long = location.longitude

#help(geopy.location)
# 34.05544, -118.33132
```
