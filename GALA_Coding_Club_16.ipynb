# Reviewing Pickle Files

Pickle files, are used for serializing and de-serializing a Python object structure. Meaning, any object in python can be pickled so that it can be saved on disk. Pickle “serializes” the object first, before writing it to file. Pickling is a way to convert a python object (list, dictionary, DataFrame, etc.) into a character stream. The idea is that this character stream contains all the information necessary to reconstruct the object in another python script, or import the data.  

Unlike other formats, pickle files are not human readable, and do not have built in security, so they are quite small and fast. A JSON file is better, if you need more security or want human readable data. In our case, we just want a way to store a smaller set of data to use for future versions of these lessons without, needing to rerun all the code on this page. 

### Importing the Shape Data Pickle files that we copied from GitHub

# Expanding Our Counting Loop

When we plotted the ZIP codes on our v1 map, we used a loop to create a dictionary, which contained the ZIP Code as the keys, and a list which contained the number of students per ZIP code, Latitude and Longitude, as the values.  We can now add the **geometry** in the same loop:

```python
zip_count = {}

for zip_code in unique_zips:
    lat = zip_df.loc[zip_code]['Lat']
    long = zip_df.loc[zip_code]['Long']
    geometry = la_geo_df.loc[str(zip_code)]['geometry']
    zip_count[str(zip_code)] = [zip_codes.count(zip_code), lat, long, geometry]
   
```

# Looking Ahead

Just like with did in v1 of the map, we could use the zip_count dictionary to plot our map.  However looking ahead there is a good reason to convert out dictionary to a pandas data frame. One, is the ease of statistical calculations.  Two is this nice visual presentation of both the tables, and geometries, when using geo pandas.

As you may recall, from the review of data structures [GALA_Coding_Club_12](https://brianarbuckle.github.io/GALA/html/GALA_Coding_Club_12.html), we can easily convert a dictionary to a data frame.

To do so, use the <code>pandas.DataFrame.from_dict()</code> method.  And to tidy things up, add column names by passing a list of names into the <code>.columns</code> *attribute*. 

Notice no parentheses? Methods have parentheses, while attributes, do not.

```python
gala_zip_df = pd.DataFrame.from_dict(zip_count, orient='index')
gala_zip_df.columns = ['Number of Students', 'Latitude', 'Longitude','geometry']
gala_zip_df.head()
```

# Interquartile Range vs Range.

When you learned how to make box plots, the whiskers usually encapsulate the total range of values.  However, in statistics, it is more common to use a multiple of the interquartile range to construct box-plots - that is a multiple of the difference between the 25% and 75% quartile. another way to say this, is the whiskers are the min and max values within 1.5 times the IQR. The dots are the outliers.

There is a whisker variable **<code>whis</code>** with a default value of 1.5, let's set it to 5:

```python
gala_zip_df.boxplot(column=['Number of Students'], vert=False, whis=5)
```
